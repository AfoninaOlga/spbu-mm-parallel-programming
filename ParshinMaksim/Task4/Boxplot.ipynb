```python
import sys
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
files = {
    'add_2000_fg.csv' : 'Add, 2000 rps, fine-grained',
    'add_2000_lazy.csv' : 'Add, 2000 rps, lazy',
    
    'all_500_lazy.csv' : 'All, 500 rps, lazy',
    'all_500_fg.csv' : 'All, 500 rps, fine-grained',
    
    'all_2000_lazy.csv' : 'All, 2000 rps, lazy',
    'all_2000_fg.csv' : 'All, 2000 rps, fine-grained',
    
    'contains_2000_fg.csv' : 'Contains, 2000 rps, fine-grained',
    'contains_2000_lazy.csv' : 'Contains, 2000 rps, lazy',
    
    'remove_2000_fg.csv' : 'Remove, 2000 rps, fine-grained',
    'remove_2000_lazy.csv' : 'Remove, 2000 rps, lazy'
}

for file, title in files.items():
    df = pd.read_csv(file, usecols = ['metric_name','metric_value'])
    
    df = df.loc[df['metric_name'].isin(['add_duration', 'remove_duration', 'contains_duration'])]
    sns.set(rc={'figure.figsize':(12, 2)})
    sns.boxplot(data=df, x='metric_value', y='metric_name', palette='magma', fliersize=2, orient='h').set(title=title, xlabel='time (ms)', ylabel='')
    
    plt.show()
```
